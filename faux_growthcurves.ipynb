```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams["font.family"] = "Helvetica"
L = [0.3, 0.5, 1.0]
k = [1, 1, 1]
x0 = [3,5,6]
colors = ["lime", "forestgreen", 'darkgreen']
x0 = 0.5
L  = [1, 1, 1]
k = [0.5, 1, 0.75]
a = [10,10,10]
b = [0.5, 1, 0.25]
labels = ['Spacer 1', 'Spacer 2', 'Spacer 3']

def sigmoid(x, k=0.1, a=0.1, b=0.1):
    # sigmoid function
    # use k to adjust the slope
    s = b/ (1 + np.exp(-(x-a)*k)) 
    return s

# un-normalised data
t = np.linspace(0, 24, 100)
# normalise the data
x = t

fig, ax = plt.subplots(1,1)
for i in range(3):
    ax.plot(x, sigmoid(x, k=k[i], a=a[i], b=b[i]), c=colors[i], label=labels[i], linewidth=5)

ax.set_xlabel('Time [H]', size=24)
ax.set_ylabel(r'Fluorescent Marker Expression [RFU]', size=12)
ax.legend(['Spacer 1', 'Spacer 2', 'Spacer 3'], fontsize=12)
plt.show()
```
